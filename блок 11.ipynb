```python
# === ЗАДАЧА 1 ===
print("=== ЗАДАЧА 1 ===")

# а) Сортировка словаря по значениям
def sort_dict_asc_desc(d):
    asc = dict(sorted(d.items(), key=lambda x: x[1]))
    desc = dict(sorted(d.items(), key=lambda x: x[1], reverse=True))
    return asc, desc

dict1 = {1:2, 3:4, 4:3, 2:1, 0:0}
asc, desc = sort_dict_asc_desc(dict1)
print(f"а) Сортировка по возрастанию: {asc}")
print(f"   Сортировка по убыванию: {desc}")

# б) Объединение словарей с суммированием
def merge_dicts_sum(d1, d2):
    result = d1.copy()
    for key, value in d2.items():
        if key in result:
            result[key] += value
        else:
            result[key] = value
    return result

dict1a = {1:2, 3:4}
dict1b = {3:5, 6:7}
merged = merge_dicts_sum(dict1a, dict1b)
print(f"б) Объединение с суммированием: {merged}")

# в) Проверка наличия ключей
def check_keys(d, keys):
    for key in keys:
        if key not in d:
            return False
    return True

dict1c = {1:2, 3:4}
keys_to_check = [1, 3]
result = check_keys(dict1c, keys_to_check)
print(f"в) Проверка ключей {keys_to_check}: {result}")
```

```python
# === ЗАДАЧА 2 ===
print("\n=== ЗАДАЧА 2 ===")

# а) Добавление нового ключа
def add_key(d, key, value):
    d[key] = value
    return d

dict2 = {1:2, 3:4}
new_dict = add_key(dict2.copy(), 5, 6)
print(f"а) Добавление ключа: {new_dict}")

# б) Уникальные значения
def unique_values(d):
    return set(d.values())

dict2b = {1:2, 3:4, 5:2}
unique_vals = unique_values(dict2b)
print(f"б) Уникальные значения: {unique_vals}")

# в) Подсчет элементов в списках-значениях
def count_list_elements(d):
    count = 0
    for value in d.values():
        if isinstance(value, list):
            count += len(value)
    return count

dict2c = {1:[2,3,4], 3:[5,6,7], 5:[8,9,0]}
total_elements = count_list_elements(dict2c)
print(f"в) Всего элементов в списках: {total_elements}")
```

```python
# === ЗАДАЧА 3 ===
print("\n=== ЗАДАЧА 3 ===")

# а) Объединение нескольких словарей
def merge_multiple_dicts(*dicts):
    result = {}
    for d in dicts:
        result.update(d)
    return result

dict3a1 = {1:2}
dict3a2 = {3:4}
dict3a3 = {5:6}
merged_multiple = merge_multiple_dicts(dict3a1, dict3a2, dict3a3)
print(f"а) Объединение словарей: {merged_multiple}")

# б) Сортировка по значениям (уже делали в задаче 1)
dict3b = {1:2, 3:4, 4:3, 2:1, 0:0}
sorted_dict = dict(sorted(dict3b.items(), key=lambda x: x[1]))
print(f"б) Сортировка по значениям: {sorted_dict}")

# в) Очистка списков в значениях
def clear_lists_in_dict(d):
    result = {}
    for key, value in d.items():
        if isinstance(value, list):
            result[key] = []
        else:
            result[key] = value
    return result

dict3c = {1:[2,3,4], 3:[5,6,7], 5:[8,9,0]}
cleared = clear_lists_in_dict(dict3c)
print(f"в) Очистка списков: {cleared}")
```
