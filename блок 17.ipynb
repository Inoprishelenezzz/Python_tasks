```python
#1
def Fact(N):
    if N==0 or N==1:
        return 1
    return N*Fact(N-1)

def Fact2(N):
    if N==0 or N==1:
        return 1
    return N*Fact2(N-2)
print(Fact(3))    
```

```python
#2
def PowerN(x, n):
    if n==0:
        return 1
    elif n>0:
        return x*PowerN(x, n-1)
    else:
        return 1/PowerN(x, -n)
print(PowerN(2, 5))
        
```

```python
#3
def SqrtK(x, k, n):
    if n==0:
        return 1
    y=SqrtK(x, k, n-1)
    if n!=0:
        return y-((y-x/y**(k-1))/k)
        
print(SqrtK(8, 3, 10))
```

```python
#4
kk=0
def FibRec(k):
    global kk
    kk+=1
    if k==1 or k==2:
        return 1
    else:
        return FibRec(k-2)+FibRec(k-1)
kkk=[5,7,10,12,15]
for n in kkk:
    kk=0
    print(FibRec(n), kk)
    
```

```python
#5
kk=0
def C(m,n):
    global kk
    kk+=1
    if m==0 or m==n:
        return 1
    else:
        return C(m, n-1)+C(m-1, n-1)
        
kkk=[1,2,3,4,5]
for m in kkk:
    kk=0
    print(C(m, n), kk)
```

```python
#6
def NOD(a,b):
    if a==0:
        return b
    return NOD(b % a, a)
        
a, b, c, d  = 48, 18, 36, 60
print(NOD(a,b))
print(NOD(a,c))
print(NOD(a,d))
```

```python
#7
def MicRec(A,N):
    if N==1:
        return A[0]
    return min(A[N-1], MicRec(A, N-1))
    
a= [[3, 1, 4, 1, 5, 9],
    [10, 2, 8, 4, 6], 
    [7, 5, 3, 9, 1]]
for i in a:
    print(MicRec(i, len(i)))
```

```python
#8
def Digits(s):
    return len([x for x in s if x.isdigit()])
    
s='dgs45hgvsj6vd'
print(Digits(s))
```

```python
#9
def Simm(s):
    if len(s)<=1:
        return True
    if s[0]!=s[-1]:
        return False
    return Simm(s[1:-1])
    
s='dgs45hgvsj6vd'
print(Simm(s))
```

```python
#11
def Root(f,a,b,e):
    if b-a<e: #проверяем достигли ли мы нужной точности
        return (a+b)/2
    c=(a+b)/2
    if f(a)*f(c)<=0: #если корень на отрезку а с то ищем корень на нем 
        return Root(f, a, c, e)
    else:
        return Root(f, c, b, e)


print(Root(lambda x:x*x - 4, 0, 3, 1e-6))
```

```python
#12
def f(n, a=0):
    if n==0:
        return a
    return f(n//10, a*10+n%10)
    
print(f(15647))
```

```python
#13
def Simm(s,i,j):
    if i-1>=j-1:
        return True
    if s[i-1]!=s[j-1]:
        return False
    return Simm(s,i-1+1,j-1-1)
    
s='dad45hgvsj6vd'
i=1
j=3
print(Simm(s,i,j))
```

```python
#14
def s(n):
    if n==0:
        return 0
    return n%10+s(n//10)

print(s(123))
```

```python
#15
def p(s, left=0, right=None):
    if right == None:
        right=len(s)-1
    if left>=right:
        return True
    if s[left]!=s[right]:
        return False
    return p(s, left+1, right-1)

if p('mom'):
    print('yes')
else:
    print('no')
```
