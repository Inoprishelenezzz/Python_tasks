```python
a = []
n = int(input('Введите количество элементов: '))

for i in range(n):
    num = int(input('Введите число: '))
    a.append(num)
ind1=a.index(max(a))
ind2=a.index(min(a))
start = min(ind1, ind2)
end = max(ind1, ind2)
sum_between = 0
for x in range(start, end + 1):
    sum_between += a[x]
print ('сумма=',sum_between)
```
