```python
import math

# 1. Середина отрезка
def midpoint(x1, y1, x2, y2):
    mid_x = (x1 + x2) / 2
    mid_y = (y1 + y2) / 2
    return mid_x, mid_y

# 2. Проверка треугольника
def can_form_triangle(a, b, c):
    # Проверяем правило треугольника
    if a + b > c and a + c > b and b + c > a:
        return True
    return False

# 3. Площадь треугольника
def area(a, b, c):
    if not can_form_triangle(a, b, c):
        return None  # Если треугольник не существует
    
    p = (a + b + c) / 2  # Полупериметр
    area = math.sqrt(p * (p - a) * (p - b) * (p - c))  # Формула Герона
    return area

print("Середина:", midpoint(0,0,4,6))
print("Треугольник существует:", can_form_triangle(3,4,5))
print("Площадь:", area(3,4,5))
```

```python
# Модуль triangle_simple.py
import math

def dist(x1,y1,x2,y2): 
    return math.sqrt((x2-x1)**2 + (y2-y1)**2)

def perimeter(ax,ay,bx,by,cx,cy):
    ab = dist(ax,ay,bx,by)
    bc = dist(bx,by,cx,cy)
    ca = dist(cx,cy,ax,ay)
    
    # Проверяем существование треугольника
    if not (ab+bc>ca and ab+ca>bc and bc+ca>ab):
        return None
    
    return ab + bc + ca

def area(ax,ay,bx,by,cx,cy):
    # Формула площади через определитель
    s = abs((bx-ax)*(cy-ay) - (cx-ax)*(by-ay))/2
    return s if s > 0 else None

# Тест

points = [(0,0), (3,0), (0,4)]
p = points[0] + points[1] + points[2]
print(p)  
print("Расстояния:", 
      dist(p[0],p[1],p[2],p[3]),
      dist(p[2],p[3],p[4],p[5]),
      dist(p[4],p[5],p[0],p[1]))
    
print("Периметр:", perimeter(*p))
print("Площадь:", area(*p))
```

```python
# 1. НОД двух чисел
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return abs(a)

# 2. Сокращение дроби
def reduce_fraction(numerator, denominator):
    if denominator == 0:
        return None, None  # Деление на ноль
    
    # Находим НОД
    divisor = gcd(numerator, denominator)
    
    # Сокращаем
    num = numerator // divisor
    den = denominator // divisor
    
    # Если знаменатель отрицательный
    if den < 0:
        num = -num
        den = -den
    
    return f"{num}/{den}"

# 3. Дробь в десятичную
def fraction_to_decimal(numerator, denominator):
    if denominator == 0:
        return None  # Деление на ноль
    
    return numerator / denominator

print("НОД(12,18):", gcd(12,18))
print("Сократить 24/36:", reduce_fraction(24,36))
print("1/2 в десятичной:", fraction_to_decimal(1,2))
```

```python
import math

# 1. Площадь круга
def area(radius):
    """
    Вычисляет площадь круга по радиусу
    Формула: S = π * r²
    """
    return math.pi * radius * radius

# 2. Длина окружности
def circumference(radius):
    """
    Вычисляет длину окружности по радиусу
    Формула: C = 2 * π * r
    """
    return 2 * math.pi * radius

# 3. Площадь сектора круга
def sector(radius, angle_degrees):
    """
    Вычисляет площадь сектора круга
    Формула: S = (π * r² * α) / 360, где α - угол в градусах
    """
    return (math.pi * radius * radius * angle_degrees) / 360

# 4. Периметр квадрата, вписанного в окружность
def square_perimeter(radius):
    """
    Вычисляет периметр квадрата, вписанного в окружность
    Сторона квадрата: a = r * √2
    Периметр: P = 4 * a
    """
    side = radius * math.sqrt(2)
    return 4 * side



r = 5
print(f"Радиус: {r}")
print(f"Площадь круга: {area(r):.2f}")
print(f"Длина окружности: {circumference(r):.2f}")
print(f"Сектор 90°: {sector(r, 90):.2f}")
print(f"Периметр впис. квадрата: {square_perimeter(r):.2f}")
```

```python
# 1. Удаление заданного символа из строки
def remove_char(text, char):
    """Удаляет все вхождения символа char из строки text"""
    result = ""
    for ch in text:
        if ch != char:
            result += ch
    return result

# 2. Замена одного символа на другой
def replace_char(text, old_char, new_char):
    """Заменяет все вхождения old_char на new_char в строке text"""
    result = ""
    for ch in text:
        if ch == old_char:
            result += new_char
        else:
            result += ch
    return result

# 3. Подсчет количества заданного символа
def count_char(text, char):
    """Считает сколько раз символ char встречается в строке text"""
    count = 0
    for ch in text:
        if ch == char:
            count += 1
    return count

text = "Hello, World!"
result1 = remove_char(text, 'o')
print(f"Удаление 'o': '{result1}'")
result2 = replace_char(text, 'o', '0')
print(f"Замена 'o' на '0': '{result2}'")
count = count_char(text, 'l')
print(f"Количество 'l': {count}")
```

```python
# 1. Печать матрицы
def print_matrix(matrix):
    for row in matrix:
        print(' '.join(map(str, row)))
    return ''
print(print_matrix([[1,2],[4,5]]))
# 2. Сложение матриц
def add_matrices(A, B):
    # Создаем пустую матрицу для результата
    result = []
    for i in range(len(A)):
        row = []
        for j in range(len(A[0])):
            # Складываем соответствующие элементы
            row.append(A[i][j] + B[i][j])
        result.append(row)
    return result

# 3. Умножение матриц
def multiply_matrices(A, B):
    # Проверяем возможность умножения
    if len(A[0]) != len(B):
        return None  # Нельзя умножить
    
    # Создаем матрицу-результат
    result = []
    for i in range(len(A)):  # строки A
        row = []
        for j in range(len(B[0])):  # столбцы B
            # Скалярное произведение строки i на столбец j
            sum_val = 0
            for k in range(len(B)):  # или len(A[0])
                sum_val += A[i][k] * B[k][j]
            row.append(sum_val)
        result.append(row)
    return result

# 4. Транспонирование матрицы
def transpose_matrix(matrix):
    # Строки становятся столбцами, столбцы - строками
    result = []
    for j in range(len(matrix[0])):  # для каждого столбца
        row = []
        for i in range(len(matrix)):  # для каждой строки
            row.append(matrix[i][j])
        result.append(row)
    return result
```

```python
# 1. Сложение векторов
def add_vectors(v1, v2):
    """Складывает два вектора одинаковой длины"""
    if len(v1) != len(v2):
        return None  # Векторы разной длины
    
    result = []
    for i in range(len(v1)):
        result.append(v1[i] + v2[i])
    return result

# 2. Умножение вектора на число
def multiply_vector(vector, scalar):
    """Умножает каждый элемент вектора на число"""
    result = []
    for value in vector:
        result.append(value * scalar)
    return result

# 3. Минимальное и максимальное значение
def min_max_vector(vector):
    """Находит минимальный и максимальный элементы вектора"""
    if not vector:
        return None, None  # Пустой вектор
    
    min_val = vector[0]
    max_val = vector[0]
    
    for value in vector:
        if value < min_val:
            min_val = value
        if value > max_val:
            max_val = value
    
    return min_val, max_val

# Пример 1: Сложение векторов
v1 = [1, 2, 3, 4]
v2 = [5, 6, 7, 8]

result = add_vectors(v1, v2)
print(f"Сложение: {v1} + {v2} = {result}")

# Пример 2: Умножение на число
scalar = 3
result = multiply_vector(v1, scalar)
print(f"Умножение: {v1} × {scalar} = {result}")

# Пример 3: Минимум и максимум
v3 = [10, -5, 8, 3, 0]
min_val, max_val = min_max_vector(v3)
print(f"Вектор: {v3}")
print(f"Минимум: {min_val}, Максимум: {max_val}")
```

```python
def remove_digits(filename):
    """Удаляет цифры из файла"""
    with open(filename, 'r') as f:
        text = f.read()
    
    # Оставляем только не-цифры
    new_text = ''.join(c for c in text if not c.isdigit())
    
    with open('clean_' + filename, 'w') as f:
        f.write(new_text)
    
    return 'clean_' + filename

def char_count(filename):
    """Считает символы"""
    with open(filename, 'r') as f:
        return len(f.read())

def line_count(filename):
    """Считает строки"""
    with open(filename, 'r') as f:
        return len(f.readlines())
```

```python
# 1. Модуль комплексного числа
def modulus(real, imag):
    """Вычисляет модуль комплексного числа a + bi"""
    return (real**2 + imag**2)**0.5

# 2. Умножение комплексных чисел
def multiply(real1, imag1, real2, imag2):
    """Умножает два комплексных числа"""
    # (a+bi)(c+di) = (ac-bd) + (ad+bc)i
    real = real1 * real2 - imag1 * imag2
    imag = real1 * imag2 + imag1 * real2
    return real, imag

# 3. Деление комплексных чисел
def divide(real1, imag1, real2, imag2):
    """Делит два комплексных числа"""
    # (a+bi)/(c+di) = [(ac+bd) + (bc-ad)i] / (c²+d²)
    denominator = real2**2 + imag2**2
    if denominator == 0:
        return None  # Деление на ноль
    
    real = (real1 * real2 + imag1 * imag2) / denominator
    imag = (imag1 * real2 - real1 * imag2) / denominator
    return real, imag
```

```python
from math import gcd #наибольший общий делитель

def power(p, q, n):
    """(p/q)^n"""
    num = p ** n
    den = q ** n
    d = gcd(num, den)
    return num//d, den//d

def equal(p1, q1, p2, q2):
    """p1/q1 == p2/q2"""
    return p1*q2 == p2*q1

def greater(p1, q1, p2, q2):
    """p1/q1 > p2/q2"""
    return p1*q2 > p2*q1

def less(p1, q1, p2, q2):
    """p1/q1 < p2/q2"""
    return p1*q2 < p2*q1

# Тест

print("(2/3)^2 =", power(2, 3, 2))
print("1/2 == 2/4?", equal(1, 2, 2, 4))
print("1/2 > 1/3?", greater(1, 2, 1, 3))
print("1/4 < 1/2?", less(1, 4, 1, 2))
```

```python
import math

# 1. Вычитание векторов
def subtract_vectors(v1, v2):
    """Вычитает вектор v2 из v1: v1 - v2"""
    x1, y1 = v1
    x2, y2 = v2
    return (x1 - x2, y1 - y2)

# 2. Скалярное произведение
def dot_product(v1, v2):
    """Скалярное произведение векторов v1 и v2"""
    x1, y1 = v1
    x2, y2 = v2
    return x1 * x2 + y1 * y2

# 3. Длина вектора
def vector_length(v):
    """Длина (модуль) вектора"""
    x, y = v
    return math.sqrt(x * x + y * y)


a = (5, 2)
b = (3, 4)

print(f"a = {a}, b = {b}")
print(f"a - b = {subtract_vectors(a, b)}")
print(f"a · b = {dot_product(a, b)}")
print(f"|a| = {vector_length(a)}")
print(f"|b| = {vector_length(b)}")
```

```python
'''
1. Вставками (как карты в руке)
Берем карту
Вставляем в нужное место среди уже отсортированных
Повторяем для всех карт

2. Обменом (пузырьковая)
Сравниваем два соседних элемента
Если левый больше правого — меняем местами
Самый большой "всплывает" как пузырек

3. Выбором
Ищем самый маленький элемент
Ставим его на первое место
Ищем следующий минимальный из оставшихся
Ставим на второе место и т.д.
'''

def insert_sort(arr):
    """Сортировка вставками """
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def bubble_sort(arr):
    """Пузырьковая сортировка"""
    n = len(arr)
    for i in range(n - 1):
        for j in range(n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def select_sort(arr):
    """Сортировка выбором"""
    n = len(arr)
    for i in range(n - 1):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

# Тест

a = [5, 3, 8, 1, 2]
print(f"Исходный: {a}")
print(f"Вставками: {insert_sort(a.copy())}")
print(f"Пузырьком: {bubble_sort(a.copy())}")
print(f"Выбором: {select_sort(a.copy())}")
```

```python
def matrix_min(m):
    """Минимальный элемент"""
    return min(min(row) for row in m) if m else None

def matrix_max(m):
    """Максимальный элемент"""
    return max(max(row) for row in m) if m else None

def matrix_sum(m):
    """Сумма всех элементов"""
    return sum(sum(row) for row in m) if m else 0

def row_sum(m, row_idx):
    """Сумма строки"""
    return sum(m[row_idx]) if m and 0 <= row_idx < len(m) else None

# Тест

mat = [[1,2,3],[4,5,6],[7,8,9]]
print(f"Матрица: {mat}")
print(f"Минимум: {matrix_min(mat)}")
print(f"Максимум: {matrix_max(mat)}")
print(f"Сумма всех: {matrix_sum(mat)}")
print(f"Сумма строки 1: {row_sum(mat, 1)}")
```

```python
def gcd(a,b):
    while b: a,b = b,a%b
    return abs(a)

def add(p1,q1,p2,q2):
    num = p1*q2 + p2*q1
    den = q1*q2
    d = gcd(num,den)
    return num//d, den//d

def sub(p1,q1,p2,q2):
    num = p1*q2 - p2*q1
    den = q1*q2
    d = gcd(num,den)
    return num//d, den//d

def mul(p1,q1,p2,q2):
    num = p1*p2
    den = q1*q2
    d = gcd(num,den)
    return num//d, den//d

# Тест

print("1/2 + 1/3 =", add(1,2,1,3))
print("3/4 - 1/4 =", sub(3,4,1,4))
print("2/3 × 3/4 =", mul(2,3,3,4))
```
