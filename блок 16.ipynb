```python
# 1
#a)
def full_name(first_name, last_name, middle_name=None):
    if middle_name:
        return f'{last_name} {first_name} {middle_name}'
    else: 
        return f'{last_name} {first_name}'
        
print(full_name('Иван', 'Петров'))

# б)
def transform_list(numbers, transform_function=None):
    if transform_function:
        return [transform_function(x) for x in numbers]
    else:
        return numbers
        
print(transform_list([1, 2, 3], transform_function=lambda x: x * 2) )

# в)
def sum_positive_numbers(*args): #собирает все аргументы в кортеж
    return sum(x for x in args if x>0)

print(sum_positive_numbers(1, -2, 3, -4, 5))

# г)
def common_keys(**kwargs):
    if not kwargs:
        return {}
    dicts=list(kwargs.values()) #извлекаем словари
    common= set(dicts[0].keys()) #извлекаем ключи из первого словаря 
    for d in dicts[1:]: #смотрим пересечение наименований ключей 
        common&=set(d.keys())
    return {key: dicts[0][key] for key in common} 

print(common_keys(dict1={'a': 1, 'b': 2}, dict2={'b': 2, 'c': 3}))

```

```python
# 2
#a)
def sort_numbers(numbers, reverse_order=None):
    if reverse_order:
        return sorted(numbers) [::-1]
    else:
        return sorted(numbers)
    
        
print(sort_numbers([4, 2, 5, 1, 3]))

# б)
def uppercase_text(text, separator=None):
    t= text.split()
    if separator:
        return separator.join(w.upper() for w in t)
    else:
        return ' '.join(w.upper() for w in t)

        
print(uppercase_text('hello world', separator=','))

# в)
def multiply_all(*w):
    p=1
    for i in w:
        p*=i
    return p
    
print(multiply_all(1, 2, 3, 4))

# г)
def filter_strings(strings, **criteria):
    result=[]
    for s in strings:
        valid = True
        if 'min_length' in criteria and len(s) < criteria['min_length']:
            valid = False
        if 'starts_with' in criteria and not s.startswith(criteria['starts_with']):
            valid = False
        if 'ends_with' in criteria and not s.endswith(criteria['ends_with']):
            valid = False
        if valid:
            result.append(s)
    return result
    
    
print(filter_strings(['яблоко', 'вишня', 'киви'], min_length=6))

```

```python
# 3
#a)
def split_string(text, delimiter=' '):
    return text.split(delimiter)
    
        
print(split_string('яблоко виноград дыня'))

# б)
def compare_elements(list1,list2,comparison_function=None):
    if comparison_function is None:
        comparison_function = lambda x, y: x == y #Эта функция проверяет равенство элементов
    result=[]
    min_len = min(len(list1), len(list2))
    for i in range(min_len):
        if comparison_function(list1[i], list2[i]):
            result.append((list1[i], list2[i]))
    return result
             
print(compare_elements([1, 2, 3], [1, 4, 3]))

# в)
def count_unique_chars(*a):
    al=set()
    for s in a:
        al.update(s)
    return len(al)
    
print(count_unique_chars('hello', 'world'))

# г)
def common_elements(*l):
    common=set(l[0])
    for ll in l[1:]:
        common&=set(ll)
    return list(common)
    
print(common_elements([1, 2, 3], [2, 3, 4], [3, 4, 5]))

```

```python
# 4
#a)
def calculate(a,b, operation='add'):
    op={'add':a+b, 'subtract':a-b, 'multiply':a*b, 'divide':a/b}
    return op.get(operation)
    
print(calculate(10, 5, operation='multiply'))

# б)
def modify_strings(strings, case=None):
    if case=='upper':
        return [s.upper() for s in strings]
    elif case=='lower':
        return [s.lower() for s in strings]
    else:
        return strings

        
print(modify_strings(['Hello', 'World'], case='upper'))

# в)
def average(*a): 
    return (sum(map(int,a)))/len(a)
    
print(average(1, 2, 3, 4, 5))

# г)
def merge_dictionaries(d, **ad):
    r=d.copy()
    for a in ad.values():
        r.update(a)
    return r
    
print(merge_dictionaries({'a': 1, 'b': 2}, dict1={'b': 3, 'c': 4}))
```

```python
# 5
#a)
def filter_items(items,condition=None):
    r=[]
    for i in items:
        if condition(i):
            r.append(i)
    if r==[]:
        r=items
    return r
    
    
print(filter_items([1, 2, 3, 4, 5], condition=lambda x: x > 3))

# б)
def merge_and_filter(list1,list2,filter_function=None):
    combined = list1 + list2
    if filter_function is None:
        return combined
    result = []
    for item in combined:
        if filter_function(item):
            result.append(item)    
    return result
    
print(merge_and_filter([1, 2], [3, 4]))

# в)
def common_in_all(*lists):
    if not lists:
        return []
    common = set(lists[0])
    for lst in lists[1:]:
        common &= set(lst)
    return list(common) 
    

print(common_in_all([1, 2, 3], [2, 3, 4], [3, 4, 5]))

# г)
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def find_primes(*numbers):
    return [n for n in numbers if is_prime(n)]
    

print(find_primes(2, 3, 4, 5, 6, 7))
```

```python
# 6
#a)
def word_count(text, separator=' '):
    words = text.split(separator)
    count_dict = {}
    for word in words:
        count_dict[word] = count_dict.get(word, 0) + 1
    return count_dict
    
print(word_count('один два один'))

# б)
def filter_numbers(numbers, exclude_multiples_of=2):
    return [n for n in numbers if n % exclude_multiples_of != 0]

        
print(filter_numbers([10, 15, 20, 25], exclude_multiples_of=5))

# в)
def merge_and_sort_keys(*di): 
    r={}
    for d in di:
        for key, value in d.items():
            if key not in r or value> r[key]:
                r[key]=value
    return sorted(r.keys(), key= lambda k: r[k], reverse=True)

print(merge_and_sort_keys({'x': 10}, {'y': 5}, {'z': 15}))

# г)
def sum_non_negative(*args):
    return sum(x for x in args if x >= 0)
    

print(sum_non_negative(1, -2, 3, -4, 5) )
```

```python
# 8
#a)
def pair_and_filter(list1, list2, filter_function=None):
    pairs = []
    for x in list1:
        for y in list2:
            if filter_function is None or filter_function(x, y):
                pairs.append((x, y))
    return pairs
    
        
print(pair_and_filter([1, 2], [3, 4]))

# б)
def concat_or_upper(strings, uppercase=False):
    result = ' '.join(strings)
    return result.upper() if uppercase else result

print(concat_or_upper(['hello', 'world']))

# в)
def filter_uppercase_strings(strings):
    return [s for s in strings if s.isupper()]
    
print(filter_uppercase_strings(['HELLO', 'World', 'BYE']))

# г)
def unique_sorted_elements(*lists):
    all_elements = set()
    for lst in lists:
        all_elements.update(lst)
    return sorted(all_elements)
    

print(unique_sorted_elements([3, 1], [2, 3], [1, 4]))
```

```python
# 9
#a)
def capitalize_words(text, separator=' '):
    words = text.split(separator)
    capitalized_words = [word.capitalize() for word in words]
    return separator.join(capitalized_words)
    
        
print(capitalize_words('python,java,c++', separator=','))

# б)
def filter_elements(list1, list2, filter_function=None):
    combined = list1 + list2
    if filter_function is None:
        return combined
    return [item for item in combined if filter_function(item)]

        
print(filter_elements([1, 2, 3], [4, 5, 6]))

# в)
def merge_dictionaries(*dicts):
    result = {}
    for d in dicts:
        result.update(d)
    return result
    

print(merge_dictionaries({'a': 1}, {'b': 2}, {'a': 3}))

# г)
def unique_keys(*dicts):
    from collections import Counter
    all_keys = []
    for d in dicts:
        all_keys.extend(d.keys())
    
    key_counts = Counter(all_keys)
    result = {}
    for d in dicts:
        for key, value in d.items():
            if key_counts[key] == 1:
                result[key] = value
    return result
    

print(unique_keys({'a': 1, 'b': 2}, {'b': 3, 'c': 4}))
```

```python
# 12
#a)
def combine_lists(list1, list2, order='first_second'):
    if order == 'second_first':
        return list2 + list1
    else:
        return list1 + list2
    
        
print(combine_lists([1, 2], [3, 4]))

# б)
def process_string(text, separator=None):
    if separator:
        words = text.split(separator)
        capitalized = [word.capitalize() for word in words]
        return separator.join(capitalized)
    else:
        return text.lower()

        
print(process_string('HELLO WORLD'))

# в)
def difference(first_list, *other_lists):
    result = []
    for item in first_list:
        in_all_others = True
        for other_list in other_lists:
            if item in other_list:
                in_all_others = False
                break
        if in_all_others:
            result.append(item)
    return result 
    

print(difference([1, 2, 3], [2, 3, 4], [3, 4, 5]))

# г)
def common_in_at_least_two(*lists):
    r={}
    for l in lists:
        unic=set(l)
        for el in unic:
            r[el]=r.get(el,0)+1 #считает сколько раз в списках встречаеься цифра 
    return [el for el, count in r.items() if count>=2]
    

print(common_in_at_least_two([1, 2], [2, 3], [2, 4]))
```

```python
# 13
#a)
def divisible_by(numbers, divisor=2):
    return [num for num in numbers if num % divisor == 0]
    
print(divisible_by([1, 2, 3, 4, 5]))

# б)
def transform_numbers(numbers, transform_function=None):
    if transform_function:
        return [transform_function(num) for num in numbers]
    else:
        return [num for num in numbers if num >= 0]

        
print(transform_numbers([1, -2, 3, -4]))

# в)
def sorted_values_by_keys(*dicts):
    a={}
    for d in dicts: #обьединяем словари
        a.update(d)

    b=[]
    sort= sorted(a.keys(), key=lambda x: int(x) if x.isdigit() else x)
    for key in sort:
        b.append(a[key])
    return b
    
    
print(sorted_values_by_keys({'2': 'b', '1': 'a'}, {'4': 'd', '3': 'c'}))

# г)
def common_keys_in_dicts(*dicts):
    
    common_keys=set(dicts[0].keys())
    for d in dicts[1:]:
        kkeys=set(d.keys())
        common_keys &= kkeys
        
    r={}
    for key in common_keys:
        for d in reversed(dicts):
            if key in d:
                r[key]=d[key]
                break
    return r
    

print(common_keys_in_dicts({'a': 1, 'b': 2}, {'b': 3, 'c': 4}))
```

```python
# 14
#a)
def merge_lists_with(list1, list2, merge_function=None):
    
    if merge_function is None:
        return list1 + list2
    
    # Используем merge_function для объединения элементов
    result = []
    min_len = min(len(list1), len(list2))
    for i in range(min_len):
        result.append(merge_function(list1[i], list2[i]))
        
    return result
    
        
print(merge_lists_with([1, 2], [3, 4]))

# б)
def extract_digits(strings, digits_only=False):
    
    if not digits_only:
        return strings
    
    result = []
    for s in strings:
        # Оставляем только цифры из каждой строки
        digits = ''.join(char for char in s if char.isdigit())
        result.append(digits)
    return result

        
print(extract_digits(['a1b2', 'c3d4']))

# в)
def filter_even_numbers(*args):
    return [num for num in args if num % 2 == 0]
    

print(filter_even_numbers(1, 2, 3, 4, 5, 6))

# г)
def sorted_positive_numbers(*args):
    positive_nums = [num for num in args if num > 0]
    return sorted(positive_nums, reverse=True)
    

print(sorted_positive_numbers(-1, 3, -5, 7, 0))
```

```python
# 15
#a)
def change_case(strings, to_upper=False):
    """
    Преобразует регистр строк и объединяет их в одну строку.
    """
    combined_text = ' '.join(strings)
    if to_upper:
        return combined_text.upper()
    else:
        return combined_text.lower()
    
        
print(change_case(['Hello', 'World']))

# б)
def filter_combined_lists(list1, list2, filter_function=None):
    """
    Фильтрует элементы из двух списков с помощью filter_function.
    """
    combined = list1 + list2
    if filter_function is None:
        return combined
    return [item for item in combined if filter_function(item)]

        
print(filter_combined_lists([2, 4], [6, 8]))

# в)
def unique_sorted_numbers(numbers):
    """
    Возвращает уникальные числа, отсортированные по возрастанию.
    """
    unique_nums = list(set(numbers))
    return sorted(unique_nums) 
    

print(unique_sorted_numbers([3, 1, 2, 3, 4, 2]))

# г)
def strings_starting_with_upper(*args):
    """
    Возвращает строки, которые начинаются с заглавной буквы.
    """
    return [s for s in args if s and s[0].isupper()]
    

print(strings_starting_with_upper('Apple', 'banana', 'Cherry'))
```
