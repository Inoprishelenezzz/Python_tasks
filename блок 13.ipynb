```python
# Задача а) - Множество латинских букв и подсчет знаков препинания
text = input("Введите текст: ")

# Множество строчных латинских букв
lowercase_letters = set()
for char in text:
    if 'a' <= char <= 'z':
        lowercase_letters.add(char)

# Подсчет знаков препинания
punctuation_count = 0
punctuation_marks = set('.,!?;:-()""')
for char in text:
    if char in punctuation_marks:
        punctuation_count += 1

print("Строчные латинские буквы:", sorted(lowercase_letters))
print("Количество знаков препинания:", punctuation_count)

# Задача б) - Проверка подмножества
A = {1, 2, 3, 4}
B = {1, 2, 3, 4, 5, 6}

# Проверяем является ли A подмножеством B
is_subset = True
for element in A:
    if element not in B:
        is_subset = False
        break

print("Множество A:", A)
print("Множество B:", B)
if is_subset:
    print("A является подмножеством B")
else:
    print("A НЕ является подмножеством B")
```

```python
ext = input("Введите строку: ")

# Создаем множество всех цифр в тексте
all_digits = {char for char in text if char in '0123456789'}

# Создаем множество цифр, которые встречаются больше одного раза
multiple_digits = {digit for digit in all_digits if text.count(digit) > 1}

# Уникальные цифры = все цифры минус те, что встречаются несколько раз
unique = all_digits - multiple_digits

print("Уникальные цифры:", sorted(unique))
print("Количество:", len(unique))

# Данные о покупках
purchases = {
    1: {'телефон', 'наушники', 'чехол'},
    2: {'ноутбук', 'мышь', 'наушники'},
    3: {'телефон', 'зарядка'},
    4: {'планшет', 'чехол', 'стилус'}
}

# Все уникальные товары
all_items = set()
# Товары, которые купили несколько пользователей
multiple_items = set()

for items in purchases.values():
    # Находим пересечение - товары, которые уже есть
    multiple_items.update(all_items & items)
    # Добавляем новые товары
    all_items.update(items)

# Товары одного пользователя
unique_items = all_items - multiple_items

print("Товары одного пользователя:", sorted(unique_items))
```

```python
# Задача а) - Уникальные буквы для трех предложений
s1 = input("Первое предложение: ")
s2 = input("Второе предложение: ")
s3 = input("Третье предложение: ")

# Буквы первого предложения, которых нет во втором и третьем
unique1 = set()
for char in s1.lower():
    if char.isalpha() and char not in s2.lower() and char not in s3.lower():
        unique1.add(char)

# Буквы второго предложения, которых нет в первом и третьем
unique2 = set()
for char in s2.lower():
    if char.isalpha() and char not in s1.lower() and char not in s3.lower():
        unique2.add(char)

# Буквы третьего предложения, которых нет в первом и втором
unique3 = set()
for char in s3.lower():
    if char.isalpha() and char not in s1.lower() and char not in s2.lower():
        unique3.add(char)

print("Уникальные буквы первого:", sorted(unique1))
print("Уникальные буквы второго:", sorted(unique2))
print("Уникальные буквы третьего:", sorted(unique3))

# Задача б) - Проверка непересекающихся множеств
A = {1, 2, 3, 4}
B = {5, 6, 7, 8}

# Простая проверка - есть ли общие элементы
has_common = False
for item in A:
    if item in B:
        has_common = True
        break

if has_common:
    print("Множества пересекаются")
else:
    print("Множества не пересекаются")
```

```python
# Задача а) - Уникальные цифры (неповторяющиеся) с множествами
number = input("Введите число: ")

# Все цифры числа
all_digits = {digit for digit in number if digit.isdigit()}

# Цифры, которые повторяются
repeated_digits = set()
for digit in all_digits:
    if number.count(digit) > 1:
        repeated_digits.add(digit)

# Уникальные цифры = все цифры минус повторяющиеся
unique_digits = all_digits - repeated_digits

print("Уникальные цифры (неповторяющиеся):", sorted(unique_digits))

# Задача б) - Строгое подмножество
A = {1, 2, 3}
B = {1, 2, 3, 4, 5}

# Проверяем что A содержится в B и B имеет дополнительные элементы
if A.issubset(B) and len(B) > len(A):
    print("A - строгое подмножество B")
else:
    print("A НЕ строгое подмножество B")
```

```python
# Задача а) - Гласные vs Согласные
text = input("Введите текст: ").lower()

# Множества гласных и согласных
vowels = {'а', 'е', 'ё', 'и', 'о', 'у', 'ы', 'э', 'ю', 'я'}
consonants = {'б', 'в', 'г', 'д', 'ж', 'з', 'й', 'к', 'л', 'м', 'н', 'п', 'р', 'с', 'т', 'ф', 'х', 'ц', 'ч', 'ш', 'щ'}

# Счетчики
vowel_count = 0
consonant_count = 0

for char in text:
    if char in vowels:
        vowel_count += 1
    elif char in consonants:
        consonant_count += 1

print(f"Гласных: {vowel_count}")
print(f"Согласных: {consonant_count}")

if vowel_count > consonant_count:
    print("Гласных больше")
elif consonant_count > vowel_count:
    print("Согласных больше")
else:
    print("Гласных и согласных поровну")
python
# Задача б) - Товары только в определенный день

# Данные о заказах по дням
orders = {
    "понедельник": [
        {'телефон', 'наушники'},
        {'ноутбук', 'мышь'}
    ],
    "вторник": [
        {'телефон', 'зарядка'},
        {'планшет', 'чехол'}
    ],
    "среда": [
        {'телефон', 'наушники'},
        {'монитор', 'клавиатура'}
    ]
}

# Выбираем день
day = input("Введите день недели: ")

if day in orders:
    # Все товары в выбранный день
    day_items = set()
    for order in orders[day]:
        day_items.update(order)
    
    # Все товары в другие дни
    other_items = set()
    for other_day, day_orders in orders.items():
        if other_day != day:
            for order in day_orders:
                other_items.update(order)
    
    # Товары только в выбранный день
    unique_items = day_items - other_items
    
    print(f"Товары только в {day}:", sorted(unique_items))
else:
    print("Такого дня нет в базе")
```
