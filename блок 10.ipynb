```python
text = "привет мир как дела"
words = text.split()
result = []

for word in words:
    if len(word) > 0:
        # Делаем первую и последнюю буквы заглавными
        new_word = word[0].upper() + word[1:-1] + word[-1].upper()
        result.append(new_word)

result_text = ' '.join(result)
```

```python
s = "abcdeflmno"  # пример строки

for i in range(1, len(s)):
    # Проверяем, является ли символ буквой и нарушает ли порядок
    if s[i].isalpha() and s[i] < s[i-1]:
        print(i + 1)  # выводим номер символа (не индекс)
        break
else:
    # Если цикл завершился без break, значит все буквы упорядочены
    print(0)
```
