```python
s=input('введите строку ')
words=s.split()
count = 0
for w in words:
    if len(w) > 0 and w[0] == w[-1]:
            count += 1
if count == 0:
    print("Не найдено слов, удов. условию")
else:
    print("Количество слов =", count)

```

```python
s=input('введите строку ')
s1=s.split()
count = 0
for w in s1:
    if "А" in w:
        count += 1
if count == 0:
   print("Нет найдено слов содержащих А")
else:
    print("Количество слов =", count)
```

```python
s = input("Введите строку: ")
count = 0
for char in s:
    if ('a' <= char <= 'z') or ('а' <= char <= 'я'):
        count += 1
print("Количество строчных латинских и русских букв",count )
```

```python
s=input('введите строку')
s1=s.split()
max=len(s1[0])
for w in s1:
    if len(w)>max:
        max=len(w)
print('длина длиного =',max)        
```

```python
s = input("Введите предложение: ")
even_chars = s[::2]
odd_chars = s[1::2]
result = even_chars + odd_chars[::-1]

print("Зашифрованная строка:", result)
```

```python
s = input("Введите строку S: ")
n = int(input("Введите число N: "))
stars = '*' * n
result = stars.join(s)
print("Результат:", result)
```
